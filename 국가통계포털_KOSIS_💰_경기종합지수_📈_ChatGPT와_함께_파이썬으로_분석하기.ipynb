#1.1 데이터 전처리

```python
obj_cols = df.describe(exclude="number").columns
obj_cols
```

```python
# "-" 값을 NaN으로 처리하고 숫자형으로 변환할 열의 리스트
# 각 열에 대해 "-" 값을 NaN으로 처리하고 숫자형으로 변환
for column in obj_cols:
    df[column] = pd.to_numeric(df[column], errors="coerce")

# 데이터 프레임에서 결측치를 제거하거나 다른 방식으로 처리할 수도 있습니다.
# 예를 들어, 결측치를 평균값이나 중간값으로 대체할 수 있습니다.
# df.fillna(df.mean(), inplace = True) # 결측치를 평균값으로 대체

# 메모리 사용량을 최적화하기 위해 float64 타입의 열만 데이터 타입을 변경합니다.
for col in df.columns:
    col_type = df[col].dtypes
    if col_type == np.float64:
        c_min = df[col].min()
        c_max = df[col].max()
        if c_min > np.finfo(np.float16).min and c_max < np.finfo(np.float16).max:
            df[col] = df[col].astype(np.float16)
        elif c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max:
            df[col] = df[col].astype(np.float32)


# 데이터프레임의 정보를 다시 확인하여 데이터 타입 및 결측치 처리 상황을 확인하세요
df.info()
```

## 상관관계
* 상관 계수의 값: 상관 계수는 -1부터 1까지의 값을 가지며, 0에 가까울수록 두 변수 간의 선형 관계가 약하거나 없음을 의미합니다. 양수 값은 양의 상관 관계(한 변수가 증가하면 다른 변수도 증가)를 의미하고, 음수 값은 음의 상관 관계(한 변수가 증가하면 다른 변수는 감소)를 의미합니

* 강도: 상관 계수의 절대값이 클수록 두 변수 간의 관계가 강하다는 것을 의미합니다. 즉, 절대값이 1에 가까울수록 강한 상관 관계를, 0에 가까울수록 약한 상관 관계를 나타냅니다.

* 방향: 상관 계수의 부호는 관계의 방향을 나타냅니다. 양수 값은 양의 상관 관계 (한 변수가 증가하면 다른 변수도 증가)를, 음수 값은 음의 상관 관계 (한 변수가 증가하면 다른 변수는 감소)를 나타냅니다.
