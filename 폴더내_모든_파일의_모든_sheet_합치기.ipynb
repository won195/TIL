```python
import pandas as pd
import os
import openpyxl
```

```python
url = '/content/drive/MyDrive/python/데이터스테이션/data/엑셀투파이썬/'
file1 = '강남점.xlsx'
url + file1
```

```python
os.listdir(url)
```

```python
df = pd.DataFrame([])
for j in os.listdir(url):
    sheet = openpyxl.load_workbook(url + j).sheetnames

    for i in sheet:
        df1 = pd.read_excel(url + j, sheet_name=i)
        df1['날짜'] = i
        df1['지점'] = j.split('.')[0]
        df = pd.concat([df,df1])
df
```

```python
df.pivot_table('판매금액', index='날짜', columns='지점', aggfunc='sum')
```
