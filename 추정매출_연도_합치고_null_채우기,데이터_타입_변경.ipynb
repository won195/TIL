```python
import pandas as pd
```

```python
df1 = pd.read_csv('/content/drive/MyDrive/상권분석/추정매출_adm.csv')
df2 = pd.read_csv('/content/drive/MyDrive/상권분석/추정매출/추정매출(2021).csv의 사본')
df3 = pd.read_csv('/content/drive/MyDrive/상권분석/추정매출/추정매출(2020).csv의 사본')
df4 = pd.read_csv('/content/drive/MyDrive/상권분석/추정매출/추정매출(2019).csv의 사본')
df5 = pd.read_csv('/content/drive/MyDrive/상권분석/추정매출/추정매출(2018).csv의 사본')
```

```python
df2 = pd.merge(df1[['lat', 'lon', 'adm_cd', 'adm_cd2', 'adm_cd8']], df2, on=['lat', 'lon'], how='right')
df3 = pd.merge(df1[['lat', 'lon', 'adm_cd', 'adm_cd2', 'adm_cd8']], df3, on=['lat', 'lon'], how='right')
df4 = pd.merge(df1[['lat', 'lon', 'adm_cd', 'adm_cd2', 'adm_cd8']], df4, on=['lat', 'lon'], how='right')
df5 = pd.merge(df1[['lat', 'lon', 'adm_cd', 'adm_cd2', 'adm_cd8']], df5, on=['lat', 'lon'], how='right')
```

```python
dfs = [df6]

for df in dfs:
    for col in df.select_dtypes(include='int64').columns:
        if not df[col].isna().any():
            try:
                if df[col].max() <= 255:
                    df[col] = df[col].astype('Int8')
                elif df[col].max() <= 65535:
                    df[col] = df[col].astype('Int16')
                elif df[col].max() <= 4294967295:
                    df[col] = df[col].astype('Int32')
            except TypeError:
                pass
```

```python
df6 = pd.concat([df1, df2, df3, df4, df5], axis=0, ignore_index=True)
```

```python
df6['adm_cd'] = df6.groupby(['lat', 'lon'])['adm_cd'].apply(lambda x: x.ffill().bfill())
df6['adm_cd2'] = df6.groupby(['lat', 'lon'])['adm_cd2'].apply(lambda x: x.ffill().bfill())
df6['adm_cd8'] = df6.groupby(['lat', 'lon'])['adm_cd8'].apply(lambda x: x.ffill().bfill())
```

```python
null_df = df6[df6[['adm_cd', 'adm_cd2', 'adm_cd8']].isnull().any(axis=1)]
null_df
```

```python
null_df['상권_코드_명'].unique()
```

```python
null_values = ['개포지하차도 북측', '대신시장', '동대문상가B동', '번동시장', '청계5가쇼핑센터(청계5가지하쇼핑센터)', '푸른터시장(기능상실)']
adm_cd_values = [1123068, 1119063, 1101067, 1108078, 1102060, 1118054]
adm_cd2_values = [1168066000, 1156063000, 1111067000, 1129078000, 1114060500, 1154563000]
adm_cd8_values = [11230680, 11190630, 11010670, 11080780, 11020600, 11180540]

for null_value, adm_cd_value, adm_cd2_value, adm_cd8_value in zip(null_values, adm_cd_values, adm_cd2_values, adm_cd8_values):
    df6.loc[df6['상권_코드_명'] == null_value, 'adm_cd'] = adm_cd_value
    df6.loc[df6['상권_코드_명'] == null_value, 'adm_cd2'] = adm_cd2_value
    df6.loc[df6['상권_코드_명'] == null_value, 'adm_cd8'] = adm_cd8_value
```
