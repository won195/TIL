```python
from bing_image_downloader import downloader

def bingImageCrawling(keyword, limit):
    downloader.download(keyword, limit=limit, output_dir='dataset', adult_filter_off=True, force_replace=False, timeout=1000)
```

```python

keyword = input("키워드 입력 : ")
limit = input("이미지 개수 : ")

bingImageCrawling(keyword, int(limit))
```
