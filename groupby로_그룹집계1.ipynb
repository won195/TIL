```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/"
                 "panda-kim/pandas/main/groupby01.csv")
df
```

```python
df.groupby(['매출일자','상품명'])[['수량','판매금액']].agg('sum')
```

```python
df.groupby(['매출일자','상품명'],as_index=False)[['수량','판매금액']].agg('sum')
```

```python
df.groupby(['매출일자','상품명'],as_index=False)[['수량','판매금액']].agg('sum').reset_index()
```
