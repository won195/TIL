```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/python/데이터스테이션/data/train.csv')
df
```

```python
df1 = df.iloc[:,[1,2,4,5]]
df1.head(4)
```

```python
df1.info()
```

```python
df1.describe()
```

```python
df1['Survived'].unique()
```

```python
df1 = df1.dropna()
df1.info()
```

```python
df1
```

```python
df2 = df1.pivot_table('Survived', index='Pclass', columns='Sex', aggfunc='mean')
df2
```

```python
import pandas_bokeh

df2.plot_bokeh(kind='bar')
```
