```python
import pandas as pd

df1 = pd.read_excel('/content/drive/MyDrive/python/데이터스테이션/data/D01EXAMPLE.xlsx', skiprows=1).set_index('Symbol Name')
df1
```

```python
df2 = df1.copy()
df2 = df2.loc[:,df2.iloc[0].isnull()].stack().reset_index()
df2.columns = ['date','stocks','price']
df2 = df2.sort_values(['stocks','date'])
df2['D+'] = df2.groupby('stocks').cumcount()
df2 = df2[df2['D+'] < 61].reindex(['stocks','D+','price','date'], axis=1)
# df2.to_clipboard(index=False)
```

```python
pdf1 = df2.pivot_table('price', index="D+", columns='stocks', aggfunc='sum')
pdf2 = pdf1 / pdf1.iloc[0]
```

```python
import pandas_bokeh
from bokeh.io import output_notebook

output_notebook()

pdf2.plot_bokeh(kind='line', legend='top_left', xlim=(pdf2.index[0], pdf2.index[-1]))
```
