```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/"
                 "panda-kim/pandas/main/apply01.csv")
df
```

```python
df.describe()
```

```python
def age_class(x):
    if 20 <= x <30:
       return '20대'
    if 30 <= x <40:
       return '30대'
    if 40 <= x <50:
       return '40대'

df['분류1'] = df['나이'].apply(age_class)
df
```

```python
df['분류1'] = df['나이'].apply(lambda x:str((x // 10)*10)+'대')
df
```

```python
df['분류2'] = df.apply(lambda x:str((x['나이']//10)*10) + x['성별'][0], axis=1)
df
```
