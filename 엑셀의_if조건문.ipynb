```python
import pandas as pd
df1 = pd.read_excel('/content/drive/MyDrive/python/데이터스테이션/data/E06EXAMPLE.xlsx', sheet_name=1)
df1
```

```python
cond1 = df1['점수'].isnull()
df1['참석여부'] = df1['점수'].mask(cond1,'X').mask(~cond1,'O')
df1
```

```python
cond2 = df1['점수'] > 70
cond3 = df1['점수'] > 90
df1['성적'] = df1 ['점수'].mask(~cond2,'C').mask(cond2,'B').mask(cond3,'A')
df1
```
