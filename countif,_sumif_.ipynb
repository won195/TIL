```python
import pandas as pd

df1 = pd.read_excel('/content/drive/MyDrive/python/데이터스테이션/data/E03EXAMPLE.xlsx', sheet_name=1)
df1
```

```python
df2 = pd.read_excel('/content/drive/MyDrive/python/데이터스테이션/data/E03EXAMPLE.xlsx', sheet_name=2)
df2
```

```python
df3 = df1.merge(df2, how='left')
df3
```

```python
pdf1 = df3.pivot_table('수량',index=['업체','메뉴'],aggfunc='count')
pdf1
```

```python
pdf2 = df3.pivot_table('가격', index='업체', aggfunc='sum')
pdf2
```
