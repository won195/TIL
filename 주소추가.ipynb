```python
df['temp'] = df['temp'].str.replace(r'\b\w+구\s*\b', '')
```

```python
import re

def extract_lat_long(geometry):
    match = re.search(r'\(\(\(([\d.]+) ([\d.]+)', geometry)
    if match:
        return float(match.group(2)), float(match.group(1))
    else:
        return None, None

df[['long', 'lat']] = df['geometry'].apply(extract_lat_long).apply(pd.Series)
```
