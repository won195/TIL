```python
df1 = fdr.DataReader("005930", "2020-01-01", "2020-12-30")
fdr.chart.plot(df1)
fdr.chart.plot(df1, title="삼성전자", color_volume_up="grey", color_volume_down="black")

df2 = pyupbit.get_ohlcv("KRW-BTC", interval="day", count=200)
df2.columns = df1.columns
fdr.chart.plot(df2, title="BTC", color_volume_up="grey", color_volume_down="black")
```
