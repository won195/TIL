```python
import pandas as pd
```

```python
df1 = pd.read_csv('/content/drive/MyDrive/python/데이터스테이션/data/Weather_Data01.csv')
print(df1.shape) # index(row) / column
df1.head()
```

```python
df1['날짜'] # 시리즈 (순열)
```

```python
df1['날짜_datetime'] = pd.to_datetime(df1['날짜'])
```

```python
df1.head(2)
```

```python
df1['연도'] = df1['날짜_datetime'].dt.year
```

```python
df1.head() 
```

```python
pd.pivot_table(data=df1, index='연도', values='출고수량', aggfunc='sum')
```

```python
df1['월'] = df1['날짜_datetime'].dt.month
df1['주차'] = df1['날짜_datetime'].dt.week
df1['일자'] = df1['날짜_datetime'].dt.day
df1['요일'] = df1['날짜_datetime'].dt.day_name()
```

```python
df1.head(2)
```

```python
pd.pivot_table(data=df1, index='월', values='출고수량', aggfunc='sum').plot(kind='bar')
```

```python
df2 = pd.read_csv('/content/drive/MyDrive/python/데이터스테이션/data/Weather_Data02.csv')
print(df2.shape) # index(row) / column
df2.head()
```

```python
pd.to_datetime(df2['배송시작일'])
```

```python
def date_function(x):
    return x[0:10]

df2['배송시작일(datetime)'] = pd.to_datetime(df2['배송시작일'].apply(date_function))
df2['배송완료일(datetime)'] = pd.to_datetime(df2['배송완료일'].apply(date_function))
```

```python
df2['배송시간'] = df2['배송완료일(datetime)'] - df2['배송시작일(datetime)']
```

```python
df2.head()
```

```python
df2['배송시간'].describe()
```

```python
pd.to_datetime(df2['주문일'])
```

```python
def date_function2(x):
    result = str(x)
    return result[0:4] +'-' +result[4:6]+'-' +result[6:8]

pd.to_datetime(df2['주문일'].apply(date_function2))
```
