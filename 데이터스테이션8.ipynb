```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rc('font', family='Malgun Gothic')
```

```python
df1= pd.read_csv('/content/drive/MyDrive/python/데이터스테이션/data/Data09.csv', encoding='cp949')
print(df1.shape)
df1.head(10)
```

```python
stack_data = df1.drop(columns=['순번','상품코드']).set_index('상품명').stack()
```

```python
df2 = pd.DataFrame(stack_data).reset_index()
```

```python
df2.head(2)
```

```python
cond1 = (df2['상품명'].isnull())
cond2 = (df2['상품명']!='합 계')
cond3 = (df2['level_1'] !=' 합계')
df3 = df2.loc[(~cond1)&(cond2)&(cond3)]
df4 = df3.rename(columns={'level_1':'점포명', 0:'판매량'})
df4
```

```python
df4['상품명'].unique()
```

```python
def func1(row):
    if 'CAT' in row:
        return 'C Type'
    elif 'STL' in row:
        return 'S Type'
    elif 'BW' in row:
        return 'B type'
    else:
        return 'Non Type'
```

```python
df4['제품군'] = df4['상품명'].apply(func1)
```

```python
df4['제품군'].value_counts()
```

```python
cond1 = (df4['제품군'] != 'Non Type')
df5 = df4.loc[cond1]
```

```python
df5.info()
```

```python
# df5['판매량'].astype(int)
```

```python
def func2(row):
    if len(row)>=4:
       result = row.split(',')
       return result[0] + result[1]
    else:
       return row

df5['판매량(int)'] = (df5['판매량'].apply(func2)).astype(int)
```

```python
df5.info()
```

```python
df6 = pd.pivot_table(data=df5, index=['점포명','제품군'],values='판매량(int)',aggfunc='sum').reset_index()
```

```python
df7 = df6.sort_values(by='판매량(int)', ascending=False)
#df7.to_csv('result_0219.csv',encoding='cp949')
```
