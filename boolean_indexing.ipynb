```python
import pandas as pd
import numpy as np
a1 = [60, 84, 80, 70, "문예부"]
a2 = [77, 62, 95, 85, "과학부"]
a3 = [61, 97, 72, np.nan, "농구부"]
a4 = [75, 65, 95, np.nan, "다도부"]
cols = ["국어", "영어", "수학", "과학", "소속"]
b1 = ["문예부", "과학부", "다도부"]
df = pd.DataFrame([a1, a2, a3, a4], index=list("ABCD"), columns=cols)
```

```python
cond1 = df['영어'] > 80
df.loc[cond1,'국어':'과학' ]
```

```python
cond1 = df['국어'] > 60
cond2 = df['수학'] > 75
df.loc[cond2 & cond1]
```

```python
df.loc[(~df['과학'].isnull()) | (df['수학'] > 90) ]
```

```python
df.loc[df['소속'] != '농구부']
```

```python
df.loc[df['소속'].isin(b1)]
```
