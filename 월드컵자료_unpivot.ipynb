```python
import pandas as pd
df1 = pd.read_excel('/content/drive/MyDrive/python/데이터스테이션/data/E14EXAMPLE.xlsx')
df2 = pd.read_excel('/content/drive/MyDrive/python/데이터스테이션/data/E14EXAMPLE.xlsx', sheet_name=2)
df1
```

```python
pdf1 = df1.loc[:,'1위':'4위'].stack().reset_index()
pdf1.columns = ['횟수','순위','국가']
pdf1 = pdf1.pivot_table('횟수',index='국가',columns='순위', aggfunc='count')
pdf1.sort_values(["1위","2위","3위","4위"] ,ascending=False).iloc[:8]
```

```python
pdf1 = df1.melt(id_vars=['개최국'], value_vars=["1위","2위","3위","4위"],var_name='순위',value_name='국가')
pdf1 = pdf1.loc[pdf1['개최국'].isin(df2['국가'])]
pdf1 = pdf1.pivot_table('개최국',index='국가',columns='순위', aggfunc='count')
pdf1.sort_values(["1위","2위","3위","4위"], ascending=False).iloc[:8]
```
