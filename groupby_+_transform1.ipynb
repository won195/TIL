```python
import pandas as pd
pd.options.display.float_format = '{:.2f}'.format
df = pd.read_csv("https://raw.githubusercontent.com/"
                 "panda-kim/pandas/main/pokemon01.csv")
df
```

```python
df['hp_m'] = df.groupby('타입')['체력'].transform('mean')
df
```

```python
cond1 = df['체력'] > df['hp_m']
df['분류'] = df['방어'].mask(cond1,'강인').mask(~cond1,'허약')
df
```

```python
df.drop('hp_m',axis=1)
```
