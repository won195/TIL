```python
import pandas as pd
a1 = ['송중기', 92, 99, 69, 75]
a2 = ['김나현', 88, 91, 87, 83]
a3 = ['권보아', 72, 86, 71, 84]
a4 = ['박효신', 81, 78, 77, 91]
a5 = ['김범수', 88, 80, 94, 91]
a6 = ["noon", "mom", "dad", "level", "river"]
a7 = ["서울", "서울", "대구", "부산", "부산"]
a8 = ["대구", "서울", "부산", "서울", "부산"]
col1 = ["이름", "국어", "영어", "수학", "과학"]
col2 = ["서울", "대구", "부산"]
df1 = pd.DataFrame([a1, a2, a3, a4, a5], columns=col1)
df2 = df1.copy()
df3 = df1.copy()
df3["과학"] = df3["과학"].astype(str)
df4 = pd.DataFrame(df1["이름"])
df4["단어"] = a6
df5 = pd.DataFrame([[1,5,10], [5,1,5], [10,5,1]], index=col2, columns=col2)*1000
df6 = pd.DataFrame([a7,a8], index=["보내는곳", "받는곳"]).T
df6 = pd.concat([df1["이름"], df6], axis=1)
df1
```

```python
df3['과학'] = df3['과학'].apply(int)
df3.info()
```

```python
df4['글자수'] = df4['단어'].apply(len)
df4
```

```python
df2['평균'] = df2.apply('mean', axis=1)
df2
```

```python
df1['수학보정'] = df1['수학'].apply(lambda x:x+1 )
df1
```

```python
df1['과학성적'] = df1['과학'].apply(lambda x:'A' if x > 90 else 'B')
df1
```

```python
df6['운임'] = df6.apply(lambda x: df5.loc[x['보내는곳'], x['받는곳']], axis=1)
df6
```
