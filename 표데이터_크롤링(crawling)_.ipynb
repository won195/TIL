```python
import pandas as pd
```

```python
df1 = pd.read_html('https://sports.news.naver.com/kbaseball/record/index?category=kbo&year=2021')[0]
```

```python
url = 'https://sports.news.naver.com/kbaseball/record/index?category=kbo&year='
year = '2021'
years = range(2015,2022)
```

```python
df = pd.DataFrame([])
for i in years:
    df1 = pd.read_html(url + year)[0]
    df1['연도'] = str(i)
    df = pd.concat([df, df1])
df
```

```python
df.pivot(values='순위',index='연도', columns='팀')
```
