```python
df2018_1 = df2018.groupby('읍면동명')[['월요일_생활인구_수', '화요일_생활인구_수', '수요일_생활인구_수', '목요일_생활인구_수', '금요일_생활인구_수', '토요일_생활인구_수', '일요일_생활인구_수']].sum()
df2018_1.info()
```

```python
df2022 = df2[(df2['시군구명']=='마포구') & (df2['기준_분기_코드']==4) & (df2['기준_년_코드']==2022)]
df2022_1 = df2022.groupby('읍면동명')[['월요일_생활인구_수', '화요일_생활인구_수', '수요일_생활인구_수', '목요일_생활인구_수', '금요일_생활인구_수', '토요일_생활인구_수', '일요일_생활인구_수']].sum()
df2022_1['주중_생활인구_수'] = df2022_1[['월요일_생활인구_수', '화요일_생활인구_수', '수요일_생활인구_수', '목요일_생활인구_수', '금요일_생활인구_수']].sum(axis=1)
df2022_1['주말_생활인구_수'] = df2022_1[['토요일_생활인구_수', '일요일_생활인구_수']].sum(axis=1)
df2022_1
```
