```python
import pandas as pd
df = pd.DataFrame([[8, 8, 2, 1, 0, 0], 
                  [7, 5, 9, 5, 7, 3], 
                  [3, 8, 1, 6, 3, 4], 
                  [7, 6, 8, 6, 4, 6], 
                  [1, 3, 0, 9, 2, 7], 
                  [3, 0, 2, 8, 2, 7]], index = range(1,7))
df.columns = ['1일차', '2일차', '3일차', '4일차', '5일차', '6일차']
df
```

```python
df['당첨'] = df.apply(lambda x: x[str(x.name) + '일차'],axis=1)
df
```
