```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_quality = pd.read_csv("https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/air_quality_no2.csv",index_col=0, parse_dates=True)
air_quality.head()
```

```python
air_quality.shape
```

```python
air_quality.plot()
```

```python
air_quality["station_paris"].plot()
plt.show()
```

```python
air_quality.plot.scatter(x="station_london", y="station_paris", alpha=0.5)

plt.show()
```

```python
[
    method_name
    for method_name in dir(air_quality.plot)
    if not method_name.startswith("_")
]
```

```python
air_quality.plot.hist(alpha=0.5, bins=50, subplots=True);
plt.show()
```

```python
air_quality.plot.kde()
```

```python
air_quality.plot.kde(subplots=True)
```

```python
air_quality.plot.box()
```

```python
air_quality.describe()
```

```python
axs = air_quality.plot.area(figsize=(12, 4), subplots=True)

plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12,6))
air_quality.plot.hist(ax=axes[0,0], alpha=0.5)
air_quality.plot.area(ax=axes[0,1])
air_quality.plot.kde(ax=axes[1,0])
air_quality.plot.scatter(x="station_paris", y="station_london", ax=axes[1,1])
```
