```python
Y = df1['target_binom']
X = df1.drop(columns=['Target', 'Error_message', 'target_binom', 'Chamber_Route',
                      'Unnamed: 0', 'Ox_Chamber', 'process', 'type',
                      'No_Die','Unnamed: 0_x', 'photo_soft_Chamber', 'process 2',
                      'Unnamed: 0_y','Etching_Chamber', 'Process 3',
                      'Chamber_Num', 'process4'])
```

```python
numeric_list = X.describe().columns.tolist()
X1 = X[numeric_list]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X1,Y, random_state=1234)

print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, Y_train)
```

```python
 model
```

```python
from sklearn.metrics import accuracy_score
```

```python
Y_train_pred = model.predict(X_train)
accuracy_score(Y_train, Y_train_pred)
```

```python
Y_test_pred = model.predict(X_test)
accuracy_score(Y_test, Y_test_pred)
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
```

```python
precision_score(Y_test, Y_test_pred)
```

```python
recall_score(Y_test, Y_test_pred)
```

```python
f1_score(Y_test, Y_test_pred)
```
