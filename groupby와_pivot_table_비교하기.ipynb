```python
import pandas as pd
pd.options.display.float_format = '{:.2f}'.format
a1 = ['송중기', '남', 1, 79, 93]
a2 = ['김나현', '여', 1, 85, 85]
a3 = ['권보아', '여', 1, 91, 78]
a4 = ['김범수', '남', 1, 97, 98]
a5 = ['이효리', '여', 2, 93, 79]
a6 = ['박효신', '남', 2, 78, 90]
a7 = ['한예슬', '여', 2, 87, 74]
a8 = ['민경훈', '남', 2, 99, 89]
col1 = ["", "성별", "반", "국어", "영어"]
df1 = pd.DataFrame([a1, a2, a3, a4, a5, a6, a7, a8], columns=col1).set_index("")
df2 = df1.copy()
df2.loc["이효리", "반"] = 1
df2["인원"] = 1
s1 = df1["국어"].copy()
df1
```

```python
df1.groupby(['성별','반'])[['국어','영어']].agg('mean')
```

```python
df1.pivot_table(['국어','영어'],index=['성별','반'],aggfunc='mean')
```

```python
df1.pivot_table('국어', index='성별', columns='반', aggfunc='mean')
```

```python
df1.groupby(['성별','반'])[['국어']].agg('mean').unstack()
```

```python
df2.groupby(['성별','반'])[['영어','인원']].agg({'영어':'mean','인원':'sum'})
```

```python
df2.groupby(['성별','반']).agg({'영어':'mean','인원':'sum'})
```

```python
df2.groupby(['성별','반'])[['국어','영어']].agg(lambda x: x.max()-x.min())
```

```python
df2.groupby(['성별','반'])[['국어','영어']].agg(lambda x:sum(x>80))
```
