```python
import pandas as pd
pd.options.display.float_format = '{:.4f}'.format
df1 = pd.read_csv("https://raw.githubusercontent.com/"
                  "panda-kim/pandas/main/shift01%EA%B0%95_df1.csv")
df1
```

```python
df1['shift'] = df1['Close'].shift(1)
df1
```

```python
df1['등락'] = (df1['Close'] - df1['shift']) / df1['shift']
df1['등락'] = df1['Close'].pct_change(1)
df1
```

```python
df1['시초등락'] = (df1['Open'] - df1['shift']) / df1['shift']
df1.drop('shift', axis=1)
```
