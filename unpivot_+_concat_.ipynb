```python
import pandas as pd

udf1 = pd.read_excel('/content/drive/MyDrive/python/데이터스테이션/data/D00EXAMPLE.xlsx').set_index('노래').stack().reset_index()
udf1.columns = ['심사자','참가자','점수']
udf1['종목'] = '노래'
udf1
```

```python
udf2 = pd.read_excel('/content/drive/MyDrive/python/데이터스테이션/data/D00EXAMPLE.xlsx', sheet_name=1).set_index('댄스').stack().reset_index()
udf2.columns = ['심사자','참가자','점수']
udf2['종목'] = '댄스'
udf2
```

```python
udf3 = pd.read_excel('/content/drive/MyDrive/python/데이터스테이션/data/D00EXAMPLE.xlsx', sheet_name=2).set_index('스타성').stack().reset_index()
udf3.columns = ['심사자','참가자','점수']
udf3['종목'] = '스타성'
udf3
```

```python
pdf1 = pd.concat([udf1,udf2,udf3])
pdf1
```

```python
pdf1 = pdf1.pivot_table('점수', index='참가자', columns='종목',aggfunc='mean')
pdf1#.to_clipboard
```
