```python
import pandas as pd
a1 = ['송중기', 1, 92, 82, 17]
a2 = ['권보아', 1, 89, 91, 17]
a3 = ['김나현', 1, 71, 91, 18]
a4 = ['박효신', 1, 90, 78, 18]
a5 = ['김선미', 1, 83, 78, 19]
a6 = ['강민경', 1, 81, 91, 19]
col = ['이름', '반', '영어', '국어', '나이']
df1 = pd.DataFrame([a1, a2, a3, a4, a5, a6], columns=col)
df2 = df1.copy()
df3 = df1.copy()
df4 = df1.copy()
```

```python
df1.rank()
```

```python
df1['영어등수'] = df1['영어'].rank(ascending=False)
df1
```

```python
df2['average'] = df2['국어'].rank(ascending=False)
df2
```

```python
df2['min'] = df2['국어'].rank(ascending=False, method='min')
df2
```

```python
df2['max'] = df2['국어'].rank(ascending=False, method='max')
df2
```

```python
df2['dense'] = df2['국어'].rank(ascending=False, method='dense')
df2
```

```python
df3['반인원'] =df3['반'].rank(method='max')
df3
```

```python
df3['순번'] =df3['반'].rank(method='first')
df3
```
