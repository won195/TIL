```python
import pandas as pd
```

```python
df1 = pd.read_csv('/content/drive/MyDrive/python/데이터스테이션/data/Data06.csv')
df1
```

```python
df2 = df1.drop(columns=['회원번호','주소-동','연령대','1회방문시구매금액(평균)'])
df2.head()
```

```python
df2.isnull().sum()
```

```python
df3 = df2.dropna()
df3
```

```python
df3['Gold_member'].value_counts()
```

```python
df3['Target'] = df3['Gold_member'].replace({'VIP':1,'normal':0})
```

```python
X = pd.get_dummies(df3.drop(columns=['Gold_member','Target']))
Y = df3['Target']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.3,random_state=2020)

print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
model = DecisionTreeClassifier()
model.fit(X_train,Y_train)
```

```python
Y_train_pred = model.predict(X_train)
Y_test_pred = model.predict(X_test)
```

```python
print(classification_report(Y_train,Y_train_pred))
```

```python
print(classification_report(Y_test,Y_test_pred))
```
