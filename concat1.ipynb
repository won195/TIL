```python
import pandas as pd

df1 = pd.read_excel('/content/drive/MyDrive/python/데이터스테이션/data/E15EXAMPLE.xlsx')
df2 = pd.read_excel('/content/drive/MyDrive/python/데이터스테이션/data/E15EXAMPLE.xlsx', sheet_name=1)
df3 = pd.read_excel('/content/drive/MyDrive/python/데이터스테이션/data/E15EXAMPLE.xlsx', sheet_name=2)
df1
```

```python
cond1 = df1['이름'].isin(df3['이름'])
df1 = df1.loc[cond1]
df1
```

```python
cond2 = df2['지명팀'] == '롯데'
df2 = df2.loc[cond2]
df2
```

```python
pd.concat([df1,df2.reindex(columns=df1.columns, fill_value=1)])
```
